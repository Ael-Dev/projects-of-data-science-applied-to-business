**¿Qué es EDA?**

Es una combinación de técnicas numéricas y de visualización que nos permite comprender las diferentes características de un conjunto de datos, sus características  y las posibles relaciones entre ellas.

Por otro lado, no confundamos el cálculo con la comprensión. Cualquiera puede calcular la desviación estándar de una característica numérica; se puede hacer por ejemplo, con el método `std()`de Pandas Series. Nuestro trabajo es usar ese número para comprender mejor sus características y el conjunto de datos.

Otro ejemplo, después de repasar algunas definiciones de distribución simétrica y asimétrica, cualquiera puede diferenciar entre una y otra. El punto no es identificar si una característica tiene una distribución simétrica o sesgada, sino usar ese conocimiento para decir algo sobre los datos, para luego poder interpretar esos hechos en el contexto del problema comercial se este tratando resolver.

Ahora ya tenemos una idea de lo que estamos tratando de realizar con EDA. ¿Pero cómo realizamos el EDA en la práctica?  Lo primero que debemos conocer es que es bastante complicado y es que nos encontraremos muchas veces probando muchas cosas a veces inútiles, confundiéndonos, y obteniendo ideas equivocadas, encontrando información contradictoria, corrigiéndote, encontrando hechos interesantes o sorprendentes, volviendo a la fase anterior para obtener más datos o para diseñar algunas características.

## **Bivariate EDA**

Una vez que hemos revisado cada una de nuestras variables de manera individual, podemos ahora revisar si existen alguna relación entre algunas de las variables.

La técnica de EDA bivariada se utiliza para explorar pares de variables y comenzar a comprender como se relacionan entre si.

¿Cuántas relaciones de pareja tendremos? Para un conjunto de $k$ variables, tendremos $\frac{k(k-1)}{2} $ distinto pares.

Si tuviéramos un conjunto de datos con 10 variables, tendriamos $\frac{10(10-1)}{2}=45 $ pares de variables para analizar. Este es un conjunto de datos muy pequeño (en término del número de variable), pero como se puede ver en la fórmula es básicamente un término cuadrático, por lo que para un conjunto de datos grandes, el número de pares aumenta rápidamente.

Por supuesto, no es necesario analizar todo los pares posibles; solo elija aquellas que sean interesante o que respondan a una pregunta en particular que pueda tener sobre el conjunto de datos, de esta manera reduciria la cantidad de pares a analizar. Por otro lado, podemos usar las libreria Pandas y Seaborn o mathplotlib, harán que nuestra tarea sea un poco menos abrumadora, que pueden realizar una gran cantidad de gráficos con pocas líneas de codigo.

Para explorar las relaciones entre pares de variables, primero , necesitamos saber que tipo de características tenemos. Dada nuestra amplia categorización de las variables numérica y categóricas, existen tres posibles de comparación:

* Dos variables numéricas.
* Dos varibales categóricas.
* Una variable numérica y una variable categórica.

Para cada caso, hay visualizaiones estándar y cálculos numéricos que utilizan los profesionales, pero eso no significa que no pueda ser creativo e inventar los suyos propios, o que estas téccnicas estándar serán suficiente para brindarle una buena información sobre lo que estamos trabajando.

### **2. variables numéricas**

Para analizar la posible relación entre dos variables numéricas, tenemos dos herramientas estándar fundamentales: 
1. El diagrama de dispersión para visualización y 
2. El coeficiente de correlación de **Pearson** como cálculo numérico.

#### **Gráfico de Dispersión**

La visualización más popular es el gráfico de dispersión, muy importante y super útil. Esto se obtiene simplemente trazando pares de puntos, cada uno de los cuales tiene el valor de una variable que determina la posición en el eje $x$ y el valor de  la otra variable determina la posición en el eje $y$.

El gráfico de dispersión es muy bueno para demostrar relación entre las variables. Si existe una relación, podemos encontrar las siguientes características:

* **Patron general:** generalmente podemos detectar la formación patrones, puede ser un patrón lineal, un patrón con curvas, patrón exponencial o uno más complicados.

* **Fuerza/Ruido:** Se refiere a la claridad con la que vemos el patrón o que tan cerca los puntos siguen este patrón. También observaremos ruido en forma de cantidad de desviación o dispersión en los puntos alrededor del patrón general. Los patrones más fuertes muestran menos ruidos y viceversas.

* **Dirección:** Si existe una relación entre las variables, notaríamos una dirección general que puede ser positivo o negativo, como sigue:

    * **Positivo:** Esto significa que ambas variables se mueven en la misma dirección. Cuando una variable aumenta, la otra también tiende a aumentar y viceversa. En este caso , el patrón va hacia arriba.
    
    * **Negativo:** Esto significa que ambas variables se mueven en la dirección opuesta, por lo que cuando una aumenta, la otra tiende a disminuir y viceversa. En este caso, el patrón general desciende.

### **Correlación de Pearson**

La herramienta estándar para comparar dos variables numéricas es el **Coeficiente de Correlación de Pearso**, comúnmente conocido como coeficiente de correlación, existen otros coeficientes de correlación, pero a como hemos dichos este es el más usado y común. Esta es un indicador numérico que muestra la fuerza de la asociación lineal entre dos variables numéricas.

La palabra clave es: **asociación o relación lineal**. Si la relación entre las características no es lineal, entonces este coeficiente de correlación podría ser engañoso, por lo que siempre es unna buena idea echar un vistazo tanto al diagrama de dispersión como el coeficiente de correlación.

Conocer las variables importantes nos ayudara a poder dar una interpretación y a un análisis más objetivo para el trabajo que estemos realizando.

 * Su valor esta siempre en el intervalo $[-1,+1]$. 
 * Su valor indica la fuerza de la relación lineal entre las variables: los valores cercanos a $-1$ indican una fuerte relación negativa entre ellos, mientras que los valores cercanos a $+1$  indican una fuerte relación positiva. Los valores cercanos a $0$ indican que no hay una relación lineal entre las características.
 * Un valor de exactamente $0$ indica la ausencia de cualquier relación lineal. Los valores exactamente $-1$ y $+1$ indican una relación perfecta entre las variables.

### **2.3.1 KNN para la imputación de valor perdido en Python usando scikit-learn**

Los valores perdidos en el conjunto de datos son un gran problema antes de que podamos entrar en el modelado. Muchos algoritmos de aprendizaje automático exigen que se imputen esos valores faltantes antes de continuar.

La forma popular (computacionalmente menos costosa) que muchos científicos de datos intentan es usar media / mediana / modo o, si se trata de una serie temporal, entonces el registro de avance o retraso.

Debe haber una mejor manera, que también es más fácil de hacer, que es la imputación de valor perdido basada en KNN ampliamente preferida.

**Sscikit-learn** es compatible de forma nativa con KNN Imputer, que ahora es oficialmente la forma más fácil y mejor (computacionalmente menos costosa) de imputar valor faltante. Es un proceso de 3 pasos para imputar / completar NaN (valores perdidos). Esta publicación es un tutorial muy breve que explica cómo imputar valores perdidos usando KNNImputer

**1. Cargar KNNImputer**

```python
from sklearn.impute import KNNImputer
```

**¿Como funciona?**

Según scikit-learn docs:

Los valores perdidos de cada muestra se imputan utilizando el valor medio de `n_neighbors` vecinos más cercanos que se encuentran en el conjunto de entrenamiento. Dos muestras están cerca si las características que no faltan en ninguna están cerca. De forma predeterminada, se utiliza una métrica de distancia euclidiana que admite valores perdidos, nan_euclidean_distances, para encontrar los vecinos más cercanos.

Crear marco de datos con valores perdidos

```python
import pandas as pd
from numpy import nan

df4 = df.copy()
```

**2. Inicializar KNNImputer**

Puede definir su propio valor n_neighbors (como es típico del algoritmo KNN).

```python
imputer = KNNImputer(n_neighbors=3)
```

**3. Imputar / completar valores perdidos**

```python
df4["duration"]=imputer.fit_transform(df4[["duration"]]).round(2)
df4["pdays"]=imputer.fit_transform(df4[["pdays"]]).round(2)
df4["previous"]=imputer.fit_transform(df4[["previous"]]).round(2)
```

```python
df4[["duration","pdays","previous"]]
```

Como puede ver arriba, ese es todo el proceso de imputación de valor faltante. Es tan simple como usar la media o la mediana, pero es más efectivo y preciso que usar un promedio simple. ### **2.8.1 Normalización estándar o escalado estándar**

El paso final antes de que los datos se pasen a los algoritmos de aprendizaje automático es escalar los datos. Puede ver que algunas columnas del conjunto de datos contienen valores pequeños, mientras que otras contienen valores muy grandes. Es mejor convertir todos los valores a una escala uniforme. Para hacerlo, puede usar la función `StandardScaler ()` del módulo `sklearn.preprocessing`.

Esto convierte las características en distribución normal estableciendo su media en $0$  y su desviación estándar en $1$. Esto se hace usando la siguiente operación, donde el valor medio de una columnan se resta de cada valor en ella, y luego el resultado se divide por el valor de la columna, valor de desviación estándar:

$$Scaler(x_i)=\frac{x_i-x_{mean} }{x_{std} } $$

A continuación podemos ver su implementación.

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
#scaling the training set
X_train = sc.fit_transform(X_train)
#scaling the test set
X_test = sc.transform (X_test)
```

## **2.9 Balanceo de datos con Smote**
Hoy en día, cualquier profesional del aprendizaje automático que trabaje con problemas de clasificación binaria debe haberse encontrado con esta situación típica de un conjunto de datos desequilibrado. Este es un escenario típico visto en muchos problemas comerciales válidos como detección de fraude, filtrado de spam, descubrimiento de enfermedades raras, detección de fallas de hardware, etc. El desequilibrio de clases es un escenario que surge cuando tenemos una distribución desigual de clases en un conjunto de datos, es decir, el número de puntos de datos en la clase negativa (clase mayoritaria) muy grande en comparación con la de la clase positiva (clase minoritaria).

Generalmente, la clase minoritaria / positiva es la clase de interés y nuestro objetivo es lograr los mejores resultados en esta clase. Si los datos desequilibrados no se tratan de antemano, esto degradará el rendimiento del modelo de clasificador. La mayoría de las predicciones corresponderán a la clase mayoritaria y tratarán las características de la clase minoritaria como ruido en los datos y las ignorarán. Esto resultará en un alto sesgo en el modelo.

### **2.9.1 Manejo de datos desequilibrados**

El remuestreo de datos es uno de los enfoques más comúnmente preferidos para lidiar con un conjunto de datos desequilibrado. En términos generales, existen dos tipos de métodos para esto i) Submuestreo ii) Sobremuestreo. En la mayoría de los casos, se prefiere el sobremuestreo a las técnicas de submuestreo. La razón es que, en el submuestreo, tendemos a eliminar instancias de los datos que pueden contener información importante.

### **2.9.2 ¿Qué es SMOTE?**

* SMOTE son las siglas de Synthetic Minority Over-sample Technique .
* Es una técnica de sobremuestreo en la que se crean nuevas observaciones sintéticas utilizando las muestras existentes de la clase minoritaria.
* Genera registros de entrenamiento virtual por interpolación lineal para la clase minoritaria.
* Después del proceso de sobremuestreo, los datos se reconstruyen y se pueden aplicar varios modelos de clasificación para los datos procesados.

### **2.9.3 ¿Cómo lo hace SMOTE?**

Selecciona muestras de la clase minoritaria que están cerca y luego traza líneas entre ellas. Los nuevos puntos de muestra se encuentran en estas líneas.

Para ser más precisos, se elige una muestra aleatoria y luego se usa un algoritmo KNN para seleccionar vecinos a los que se dibujan las líneas. Con este procedimiento, puede crear tantas muestras sintéticas como necesite. Esto hace que SMOTE sea perfecto para conjuntos de datos de todos los tamaños.

El único inconveniente real es que los ejemplos sintéticos se crean sin "consultar" a la clase mayoritaria. Esto podría resultar en muestras superpuestas en ambas clases.

## **2.10 Técnica de remuestreo**
Una técnica ampliamente adoptada para tratar con conjuntos de datos altamente desequilibrados se llama remuestreo. Consiste en eliminar muestras de la clase mayoritaria (submuestreo) y / o agregar más ejemplos de la clase minoritaria (sobremuestreo).

```python
#@title
from IPython.display import Image
Image("/content/drive/MyDrive/Curso de ML 3 version/Sesión N° 02/imagen/under_over-sampling.JPG")
```

A pesar de la ventaja de equilibrar las clases, estas técnicas también tienen sus debilidades (no hay almuerzo gratis).

La implementación más simple del sobremuestreo es duplicar registros aleatorios de la clase minoritaria, lo que puede causar sobrepesca.

En el submuestreo , la técnica más simple consiste en eliminar registros aleatorios de la clase mayoritaria, lo que puede provocar la pérdida de información.

### 2.10.1 **Submuestreo aleatorio (Random Under-Sampling)**
El submuestreo se puede definir como eliminar algunas observaciones de la clase mayoritaria . Esto se hace hasta que se equilibre la clase mayoritaria y minoritaria.

Significa que la clase minoritaria será la misma cantidad (1 a 1) que la clase mayoritaria, la clase minoritaria copiará sus filas.

El submuestreo puede ser una buena opción cuando tiene una tonelada de datos, piense en millones de filas. Pero un inconveniente del submuestreo es que estamos eliminando información que puede ser valiosa.

```python
from collections import  Counter
from imblearn.under_sampling import RandomUnderSampler


X = datos.drop("y_yes", axis=1)
y = datos["y_yes"]
```

```python
rus=RandomUnderSampler(sampling_strategy=1)
x_rus, y_rus= rus.fit_resample(X,y)
```

```python
print('Forma del conjunto de datos original:', Counter(y))
print('Forma del conjunto de datos con Resample ', Counter(y_rus))
```

```python
# plot the no of 1's and 0's
g = sns.countplot(y_rus)
g.set_xticklabels(['0', '1'])
plt.title(" Random Under-Sampling")
plt.show()
```

Ajustamos la estrategia de muestreo como 1. Signinfica que la clase mayoritaria será la misma cantidad que la clase minoritaria (1 a 1), la clase mayoritoria perderá filas, esto lo podemos verificar con los recuentos de `y_res`.

Ahora ya tenemos listo nuestro datos para entrenar nuestro modelo. 

### **2.10.2 Sobremuestreo aleatorio(Random Over-Sampling)**


El sobremuestreo se puede definir como agregar más copias a la clase minoritaria. El sobremuestreo puede ser una buena opción cuando no tiene una gran cantidad de datos con los que trabajar.

Establecemos la estrategia de muestreo en 1. Significa que la clase mayoritaria será la misma cantidad (1 a 1) que la clase minoritaria, la clase mayoritaria copiará sus filas.

Una desventaja a tener en cuenta al realizar un submuestreo es que puede provocar un sobreajuste y una mala generalización en su conjunto de prueba.

```python
g = sns.countplot(y)
g.set_xticklabels(['0', '1'])
plt.title("Datos  Desbalanceado")
plt.show();
```

```python
from imblearn.over_sampling import RandomOverSampler
ros=RandomOverSampler(sampling_strategy=1)

x_ros, y_ros= ros.fit_resample(X,y)
```

```python
print('Forma del conjunto de datos original:', Counter(y))
print('Forma del conjunto de datos con Resample ', Counter(y_ros))
```

```python
g = sns.countplot(y_ros)
g.set_xticklabels(['0', '1'])
plt.title("Datos Balanceados")
plt.show();
```
