## **Bivariate EDA**

Una vez que hemos revisado cada una de nuestras variables de manera individual, podemos ahora revisar si existen alguna relacin entre algunas de las variables.

La tcnica de EDA bivariada se utiliza para explorar pares de variables y comenzar a comprender como se relacionan entre si.

Cuntas relaciones de pareja tendremos? Para un conjunto de $k$ variables, tendremos $\frac{k(k-1)}{2} $ distinto pares.

Si tuviramos un conjunto de datos con 10 variables, tendriamos $\frac{10(10-1)}{2}=45 $ pares de variables para analizar. Este es un conjunto de datos muy pequeo (en trmino del nmero de variable), pero como se puede ver en la frmula es bsicamente un trmino cuadrtico, por lo que para un conjunto de datos grandes, el nmero de pares aumenta rpidamente.

Por supuesto, no es necesario analizar todo los pares posibles; solo elija aquellas que sean interesante o que respondan a una pregunta en particular que pueda tener sobre el conjunto de datos, de esta manera reduciria la cantidad de pares a analizar. Por otro lado, podemos usar las libreria Pandas y Seaborn o mathplotlib, harn que nuestra tarea sea un poco menos abrumadora, que pueden realizar una gran cantidad de grficos con pocas lneas de codigo.

Para explorar las relaciones entre pares de variables, primero , necesitamos saber que tipo de caractersticas tenemos. Dada nuestra amplia categorizacin de las variables numrica y categricas, existen tres posibles de comparacin:

* Dos variables numricas.

Para analizar la posible relacin entre dos variables numricas, tenemos dos herramientas estndar fundamentales: 
1. El diagrama de dispersin para visualizacin y

#### **Grfico de Dispersin**

La visualizacin ms popular es el grfico de dispersin, muy importante y super til. Esto se obtiene simplemente trazando pares de puntos, cada uno de los cuales tiene el valor de una variable que determina la posicin en el eje $x$ y el valor de  la otra variable determina la posicin en el eje $y$.

El grfico de dispersin es muy bueno para demostrar relacin entre las variables. Si existe una relacin, podemos encontrar las siguientes caractersticas:

* **Patron general:** generalmente podemos detectar la formacin patrones, puede ser un patrn lineal, un patrn con curvas, patrn exponencial o uno ms complicados.

* **Fuerza/Ruido:** Se refiere a la claridad con la que vemos el patrn o que tan cerca los puntos siguen este patrn. Tambin observaremos ruido en forma de cantidad de desviacin o dispersin en los puntos alrededor del patrn general. Los patrones ms fuertes muestran menos ruidos y viceversas.

* **Direccin:** Si existe una relacin entre las variables, notaramos una direccin general que puede ser positivo o negativo, como sigue:

    * **Positivo:** Esto significa que ambas variables se mueven en la misma direccin. Cuando una variable aumenta, la otra tambin tiende a aumentar y viceversa. En este caso , el patrn va hacia arriba.
    
    * **Negativo:** Esto significa que ambas variables se mueven en la direccin opuesta, por lo que cuando una aumenta, la otra tiende a disminuir y viceversa. En este caso, el patrn general desciende.

### **Correlacin de Pearson**

La herramienta estndar para comparar dos variables numricas es el **Coeficiente de Correlacin de Pearso**, comnmente conocido como coeficiente de correlacin, existen otros coeficientes de correlacin, pero a como hemos dichos este es el ms usado y comn. Esta es un indicador numrico que muestra la fuerza de la asociacin lineal entre dos variables numricas.

La palabra clave es: **asociacin o relacin lineal**. Si la relacin entre las caractersticas no es lineal, entonces este coeficiente de correlacin podra ser engaoso, por lo que siempre es unna buena idea echar un vistazo tanto al diagrama de dispersin como el coeficiente de correlacin.

Conocer las variables importantes nos ayudara a poder dar una interpretacin y a un anlisis ms objetivo para el trabajo que estemos realizando.

 * Su valor esta siempre en el intervalo $[-1,+1]$. 
 * Su valor indica la fuerza de la relacin lineal entre las variables: los valores cercanos a $-1$ indican una fuerte relacin negativa entre ellos, mientras que los valores cercanos a $+1$  indican una fuerte relacin positiva. Los valores cercanos a $0$ indican que no hay una relacin lineal entre las caractersticas.
 * Un valor de exactamente $0$ indica la ausencia de cualquier relacin lineal. Los valores exactamente $-1$ y $+1$ indican una relacin perfecta entre las variables.

## **2.9 Balanceo de datos con Smote**
Hoy en da, cualquier profesional del aprendizaje automtico que trabaje con problemas de clasificacin binaria debe haberse encontrado con esta situacin tpica de un conjunto de datos desequilibrado. Este es un escenario tpico visto en muchos problemas comerciales vlidos como deteccin de fraude, filtrado de spam, descubrimiento de enfermedades raras, deteccin de fallas de hardware, etc. El desequilibrio de clases es un escenario que surge cuando tenemos una distribucin desigual de clases en un conjunto de datos, es decir, el nmero de puntos de datos en la clase negativa (clase mayoritaria) muy grande en comparacin con la de la clase positiva (clase minoritaria).

Generalmente, la clase minoritaria / positiva es la clase de inters y nuestro objetivo es lograr los mejores resultados en esta clase. Si los datos desequilibrados no se tratan de antemano, esto degradar el rendimiento del modelo de clasificador. La mayora de las predicciones correspondern a la clase mayoritaria y tratarn las caractersticas de la clase minoritaria como ruido en los datos y las ignorarn. Esto resultar en un alto sesgo en el modelo.

### **2.9.1 Manejo de datos desequilibrados**

El remuestreo de datos es uno de los enfoques ms comnmente preferidos para lidiar con un conjunto de datos desequilibrado. En trminos generales, existen dos tipos de mtodos para esto i) Submuestreo ii) Sobremuestreo. En la mayora de los casos, se prefiere el sobremuestreo a las tcnicas de submuestreo. La razn es que, en el submuestreo, tendemos a eliminar instancias de los datos que pueden contener informacin importante.

### **2.9.2 Qu es SMOTE?**

* SMOTE son las siglas de Synthetic Minority Over-sample Technique .
* Es una tcnica de sobremuestreo en la que se crean nuevas observaciones sintticas utilizando las muestras existentes de la clase minoritaria.
* Genera registros de entrenamiento virtual por interpolacin lineal para la clase minoritaria.
* Despus del proceso de sobremuestreo, los datos se reconstruyen y se pueden aplicar varios modelos de clasificacin para los datos procesados.

### **2.9.3 Cmo lo hace SMOTE?**

Selecciona muestras de la clase minoritaria que estn cerca y luego traza lneas entre ellas. Los nuevos puntos de muestra se encuentran en estas lneas.

Para ser ms precisos, se elige una muestra aleatoria y luego se usa un algoritmo KNN para seleccionar vecinos a los que se dibujan las lneas. Con este procedimiento, puede crear tantas muestras sintticas como necesite. Esto hace que SMOTE sea perfecto para conjuntos de datos de todos los tamaos.

El nico inconveniente real es que los ejemplos sintticos se crean sin "consultar" a la clase mayoritaria. Esto podra resultar en muestras superpuestas en ambas clases.

## **2.10 Tcnica de remuestreo**
Una tcnica ampliamente adoptada para tratar con conjuntos de datos altamente desequilibrados se llama remuestreo. Consiste en eliminar muestras de la clase mayoritaria (submuestreo) y / o agregar ms ejemplos de la clase minoritaria (sobremuestreo).

```python
#@title
from IPython.display import Image
Image("/content/drive/MyDrive/Curso de ML 3 version/Sesion N 02/imagen/under_over-sampling.JPG")
```

A pesar de la ventaja de equilibrar las clases, estas tcnicas tambin tienen sus debilidades (no hay almuerzo gratis).

La implementacin ms simple del sobremuestreo es duplicar registros aleatorios de la clase minoritaria, lo que puede causar sobrepesca.

En el submuestreo , la tcnica ms simple consiste en eliminar registros aleatorios de la clase mayoritaria, lo que puede provocar la prdida de informacin.

### 2.10.1 **Submuestreo aleatorio (Random Under-Sampling)**
El submuestreo se puede definir como eliminar algunas observaciones de la clase mayoritaria . Esto se hace hasta que se equilibre la clase mayoritaria y minoritaria.

Significa que la clase minoritaria ser la misma cantidad (1 a 1) que la clase mayoritaria, la clase minoritaria copiar sus filas.

El submuestreo puede ser una buena opcin cuando tiene una tonelada de datos, piense en millones de filas. Pero un inconveniente del submuestreo es que estamos eliminando informacin que puede ser valiosa.

```python
from collections import  Counter
from imblearn.under_sampling import RandomUnderSampler


X = datos.drop("y_yes", axis=1)
y = datos["y_yes"]
```

```python
rus=RandomUnderSampler(sampling_strategy=1)
x_rus, y_rus= rus.fit_resample(X,y)
```

```python
print('Forma del conjunto de datos original:', Counter(y))
print('Forma del conjunto de datos con Resample ', Counter(y_rus))
```

```python
# plot the no of 1's and 0's
g = sns.countplot(y_rus)
g.set_xticklabels(['0', '1'])
plt.title(" Random Under-Sampling")
plt.show()
```

Ajustamos la estrategia de muestreo como 1. Signinfica que la clase mayoritaria ser la misma cantidad que la clase minoritaria (1 a 1), la clase mayoritoria perder filas, esto lo podemos verificar con los recuentos de `y_res`.

Ahora ya tenemos listo nuestro datos para entrenar nuestro modelo. 

### **2.10.2 Sobremuestreo aleatorio(Random Over-Sampling)**


El sobremuestreo se puede definir como agregar ms copias a la clase minoritaria. El sobremuestreo puede ser una buena opcin cuando no tiene una gran cantidad de datos con los que trabajar.

Establecemos la estrategia de muestreo en 1. Significa que la clase mayoritaria ser la misma cantidad (1 a 1) que la clase minoritaria, la clase mayoritaria copiar sus filas.

Una desventaja a tener en cuenta al realizar un submuestreo es que puede provocar un sobreajuste y una mala generalizacin en su conjunto de prueba.

```python
g = sns.countplot(y)
g.set_xticklabels(['0', '1'])
plt.title("Datos  Desbalanceado")
plt.show();
```

```python
from imblearn.over_sampling import RandomOverSampler
ros=RandomOverSampler(sampling_strategy=1)

x_ros, y_ros= ros.fit_resample(X,y)
```
