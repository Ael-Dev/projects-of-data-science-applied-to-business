TRATAMIENTO DE DATOS OUTLIERS MEDIANTE LA TECNICA IQR

````python
def replace_outliers_with_iqr(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[col] = df[col].mask(df[col] < lower_bound, lower_bound)
    df[col] = df[col].mask(df[col] > upper_bound, upper_bound)
    return df
````

````python
from sklearn import set_config
from sklearn.compose import make_column_transformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, StandardScaler
from sklearn.pipeline import make_pipeline

col_numericas = X2.select_dtypes(include='number').columns
col_label_encoding = ["Nivel_Educativo"]
col_categoricas = X2.select_dtypes(include='object').columns.difference(col_label_encoding).tolist()

columnas_numericas_pipe = make_pipeline(
    SimpleImputer(),
    StandardScaler()
)

# Usar OrdinalEncoder en lugar de LabelEncoder
columnas_categoricas_LE_pipe = make_pipeline(
    SimpleImputer(strategy='most_frequent'),
    OrdinalEncoder()
)
columnas_categoricas_OHE_pipe = make_pipeline(
    SimpleImputer(strategy='most_frequent'),
    OneHotEncoder(sparse_output=True, handle_unknown='ignore', drop='first')
)

pipeline = make_column_transformer(
    (columnas_numericas_pipe, col_numericas),
    (columnas_categoricas_LE_pipe, col_label_encoding),
    (columnas_categoricas_OHE_pipe, col_categoricas)
)

set_config(display='diagram')
pipeline # muestra el diagrama del pipeline
````

## **<font color='Purple'>Evaluación</font>**

````python
from sklearn.metrics import confusion_matrix,classification_report, precision_score, accuracy_score, recall_score

def graficar_matriz_confusion(modelos, x, y_true):
    for nombre, modelo in modelos.items():
        #print(f'{nombre}')
        # predecir
        y_pred = modelo.predict(x)
        # obtener la matriz de confusion
        cm = confusion_matrix(y_true, y_pred)

        # Crear una figura con matplotlib
        plt.figure(figsize=(3, 3))
        # Mostrar la matriz de confusión con seaborn
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
        plt.xlabel('Etiqueta predicha')
        plt.ylabel('Etiqueta verdadera')
        plt.title(f'Matriz de confusión del modelo {nombre}', size=14)
        plt.show()

        # save the model
        #modelos_entrenados[nombre] = modelo
        #print(f'{nombre} entrenado!')



def reporte_modelos(modelos, x, y_true):
    resultados = {}
    for nombre, modelo in modelos.items():
        print(f'{nombre}')
        y_pred = modelo.predict(x)

        # evaluar modelos
        precision = precision_score(y_true, y_pred)
        accuracy = accuracy_score(y_true, y_pred)
        recall = recall_score(y_true, y_pred)
        # guardar resultados
        resultados[nombre] = [precision, accuracy, recall]

        print(classification_report(y_true, y_pred))
    return resultados
````
